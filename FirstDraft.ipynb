1. Is there a correlation between the Judging Score and an interest in sports?

```python
# Create a helper column where Interest is 1 if the value is "Sports" and 0 otherwise
df['Interest_Sports'] = df['Interest'].apply(lambda x: 1 if x == "Sports" else 0)

# Filter the dataset to only include people interested in sports (Interest_Sports == 1)
df_sports = df[df['Interest_Sports'] == 1]

# Define the predictors (Judging Score, Introversion Score, Sensing Score, Thinking Score)
X = df_sports[['Judging Score', 'Introversion Score', 'Sensing Score', 'Thinking Score']]

# Add a constant for intercept
X = sm.add_constant(X)

# Define the target variable (Interest_Sports)
y = df_sports['Interest_Sports']

# Fit the logistic regression model
model = sm.Logit(y, X)
result = model.fit()

# Print the summary table
print(result.summary())
```

```python
sports_interest = df[df['Interest'] == "Sports"]

# Count occurrences of each personality type among those interested in sports
sports_personality_counts = sports_interest['Personality'].value_counts()

# Count occurrences of each personality type in the overall data
overall_personality_counts = df['Personality'].value_counts()

# Calculate the percentage of each personality type interested in sports
personality_sports_percentage = (sports_personality_counts / overall_personality_counts) * 100

personality_sports_percentage_sorted = personality_sports_percentage.sort_values(ascending=False)

print("Personality Types and Their Likelihood of Being Interested in Sports (%):")
print(personality_sports_percentage_sorted)
```

```python
# Load your data (uncomment if reading from CSV)
# df = pd.read_csv("your_data.csv")

# Check the column names
print(df.columns)

# If Education column is binary (1 for higher education, 0 for not higher education), use it directly
df['Higher_Education'] = df['Education']  # Assuming Education is already 1 for higher education, 0 for not

# Define the predictor (Higher_Education) and target (Thinking Score)
X = df[['Higher_Education']]
y = df['Thinking Score']

# Add a constant for the intercept
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X)
result = model.fit()

# Print the summary to check the results
print(result.summary())
```
